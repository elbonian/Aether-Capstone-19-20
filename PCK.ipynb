```python
planets = ["Sun","Mercury","Venus","Earth","Mars","Jupiter","Saturn","Uranus","Neptune","Pluto"]

satellites = {"Moon":["Moon"],"Mars":["Phobos","Deimos"],
              "Jupiter":["Io","Europa","Ganymede","Callisto","Amalthea","Himalia","Elara","Pasiphae",
                         "Sinope","Lysithea","Carme","Ananke","Leda","Thebe","Adrastea","Metis"],
              "Saturn":["Mimas","Enceladus","Tethys","Dione","Rhea","Titan","Hyperion","Iapetus","Phoebe",
                        "Janus","Epimetheus","Helene","Telesto","Calypso","Atlas","Prometheus","Pandora",
                        "Pan","Methone","Pallene","Polydeuces","Daphnis","Anthe"],
              "Uranus":["Ariel","Umbriel","Titania","Oberon","Miranda","Cordelia","Ophelia","Bianca",
                        "Cressida","Desdemona","Juliet","Portia","Rosalind","Belinda","Puck"],
              "Neptune":["Triton","Nereid","Naiad","Thalassa","Despina","Galatea","Larissa","Proteus"],
              "Pluto":["Charon"]}
```

```python
print("RADII and MASS of Sun and planets")
print("-----------------------------------")
for planet in planets:
    print("{} RADII {}".format(planet,spiceypy.bodvrd(planet,"RADII",3)[1]))
    print("{} MASS {}\n".format(planet,spiceypy.bodvrd(planet,"GM",1)[1]))
```

```python
print("RADII and MASS of satellites")
print("------------------------------")
for planet in satellites.keys():
    print("Satellites of {}:\n".format(planet))
    for satellite in satellites.get(planet):
        try:
            print("{} RADII {}".format(satellite,spiceypy.bodvrd(satellite,"RADII",3)[1]))
            print("{} MASS {}\n".format(satellite,spiceypy.bodvrd(satellite,"GM",1)[1]))
        except:
            print("VALUE ERROR\n")
```
